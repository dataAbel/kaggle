```python
def get_ks_plot(score):
    score_good=score[score['y6']==0]
    score_bad=score[score['y6']==1]
    score_good_table=pd.crosstab(score_good['模型分数'],score_good['y6'])
    score_good_table['模型分数']=score_good_table.index
    score_good_table.columns=['good','模型分数']
    score_bad_table=pd.crosstab(score_bad['模型分数'],score_bad['y6'])
    score_bad_table['模型分数']=score_bad_table.index
    score_bad_table.columns=['bad','模型分数']
    score_good_table.index.name = '分数'
    score_bad_table.index.name = '分数'

    score_table = pd.merge(score_good_table,score_bad_table,how='outer',on='模型分数')
    score_table=score_table.fillna(0)
    score_table=score_table.sort_values('模型分数',ascending=True)
    score_table['cum_sum_good']=score_table.good.cumsum()/sum(score_table.good)
    score_table['cum_sum_bad']=score_table.bad.cumsum()/sum(score_table.bad)
    ks_value=max(abs(score_table['cum_sum_good']-score_table['cum_sum_bad']))
    
    print(ks_value)
    #plt.xlim(xmax=max(score_table['totalscore']),xmin=min(score_table['totalscore']))
    #plt.ylim(ymax=1,ymin=0)
    plt.scatter(x=score_table['模型分数'],y=score_table['cum_sum_good'],color='blue',s=10)
    plt.scatter(x=score_table['模型分数'],y=score_table['cum_sum_bad'],color='red',s=10)
    plt.xlabel('模型分数',fontsize=16)
    plt.ylabel('cum_sum',fontsize=16)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.title('KS=%s' %np.round(ks_value,3),fontsize=16)
    plt.show()
    return score_table

dataset = pd.read_csv(r'E:\cga\model\多头借贷卡\output\train_data_score.csv', encoding='gb18030')
dataset = dataset.rename(columns={'total_score': '模型分数', 'y': 'y6'})[['模型分数', 'y6', 'APP_ID']]
get_ks_plot(dataset)
# dataset.sample(10)
```
